```python
import pandas as pd

from connect import database
```

```python
df = database
df
```

# Total each field

```python
# Total Brands
total_brand = len(df.groupby(['brand_name'])['brand_name'])
total_brand
```

```python
# Total Product Name
total_product = df.groupby(['brand_name','name'])['name'].size().reset_index(name='count')
total_product
```

```python
# Total Product Name
total_product_by_brand = df.groupby(['brand_name'])['brand_name'].size().reset_index(name='count')
total_product_by_brand
```

```python
# Total Product by Seller
total_product_by_seller = df.groupby(['seller'])['seller'].size().reset_index(name='count').sort_values(ascending=False, by='count')
total_product_by_seller
```

```python
# Total Product is genuine
total_product_is_genuine = df.groupby(['is_genuine']).size()
total_product_is_genuine
```

```python
# Total Product is Second Hand
total_product_is_used = df.groupby(['is_used']).size()
total_product_is_used
```

```python
# Total Product is Imported
total_product_is_imported = df.groupby(['is_imported_goods']).size()
total_product_is_imported
```

```python
for i, row in df.iterrows():
    if row['original_price'] < row['sell_price']:
        df.loc[i, 'sell_price'] = row['original_price']
df
```

```python
df_rating_sum = df.groupby(['brand_name']).sum()
df_rating_sum.sort_values(by="rating_5",ascending=False)
```

```python
df['discount'] = df['original_price'] - df['sell_price']
df['discount_percent'] = round(df['discount']/df['original_price'],2)
df
```

```python
discount_max = max(df['discount_percent'])
discount_max
```

```python
discount_min = min(df['discount_percent'])
discount_min
```

```python
# Discount
discount_group = df.groupby(['discount_percent']).size()
discount_group
```

```python
# Color
color_group = df.groupby(['color']).size().reset_index(name='count')
color_group
```
